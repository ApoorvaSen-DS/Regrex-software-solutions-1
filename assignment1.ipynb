## Pattern 6

```python
rows = int(input("Enter the number of rows:"))
def fun1(rows):
    for i in range(0,rows):
        for j in range(0,i+1):
            print(number(i,j), end=' ')
        print(" ")
        
def number(n,m):
    num=1
    if m>n-m:
        m=n-m
    for i in range(0,m):
        num=num*(n-i)
        num=num//(i+1)
    return num
fun1(rows)
```

## Pattern 9

```python
rows = int(input("Enter the number of rows: "))    
k = 2 * rows - 2 
for i in range(rows, -1, -1):    
    for j in range(k, 0, -1):  
        print(end=" ")  
    k = k + 1  
    for j in range(0, i + 1):  
        print("*", end=" ")  
    print("") 
```

## Pattern 15

```python
rows = int(input("Enter the number of rows:"))
print("*" * rows, end='\n')
i=(rows//2)-1
j=2
while i!=0:
    while j<=(rows-2):
        print("*" * i,end="")
        print("_" * j,end="")
        print("*" * i,end="\n")
        i-=1
        j+=2
```
