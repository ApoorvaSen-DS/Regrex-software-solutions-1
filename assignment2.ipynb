```python
# Q4. write a function to return a list of prime factors of a given number.
num = int(input("Enter the number: "))

def prime_factors(number):
    l=[]
    for i in range(2,(number+1)):
        if(number%i==0):
            flag = 1
            for j in range(2, (i//2+1)):
                if(i%j==0):
                    flag=0
                    break
                
            if(flag == 1):
                l.append(i)
    print(l)
prime_factors(num)
```
